````python
torch.manual_seed(43052)
x1,_ = torch.randn(100).sort(); x1 = x1.reshape(-1,1)
x2,_ = torch.randn(100).sort(); x2 = x2.reshape(-1,1)
eps = torch.randn((100,1))*0.5
y = 2.5+ 4*x1 -2*x2 + eps 
````
