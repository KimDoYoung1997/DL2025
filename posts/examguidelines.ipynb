# 시험 안내

## 시험 시간

- 총 40분

## 제출물

- 답안지 (`.ipynb`)
- 시험 장면 영상 촬영본

## 제출 방법

- 답안지와 영상촬영본 모두 LMS에 업로드

## 영상 촬영 방법

### Windows 사용자 (Zoom 활용)

1. Zoom 로그인하여 혼자 새 회의 시작 (비디오, 마이크 셋팅 불필요) 
2. 화면 공유 시작 $\to$ 전체화면 공유 
3. 녹화 버튼 클릭
4. 시험 종료 후 화면공유 종료 $\to$ 녹화종료 $\to$ 영상을 이 컴퓨터에 저장
   - 기본 저장 위치: 내 PC > 문서 > Zoom 폴더

※ Zoom 무료 계정은 40분 제한이 있으니, 필요한 경우 나누어 촬영

### Mac 사용자 (기본 기능 활용)

1. Command + Shift + 5 키 입력
2. 화면 하단 도구에서 전체 화면 지정 후 녹화 시작
3. 시험 종료 후 메뉴 막대에서 정지 버튼 클릭
4. 영상은 기본적으로 바탕화면에 저장됨

> 기타 OBS Studio 등 녹화 프로그램을 활용해도 무방함. 녹화본은 반드시 시험 시작부터 종료까지 전 과정을 포함해야 하며 LMS 업로드 마감 시간 내 제출할 것
